```python
import csv
include_ids = set()
include_csv = "./raw_data/nonprocedural_filteredocr_speechids.csv"
with open(include_csv, mode='r', encoding='utf-8') as f_exc:
    reader = csv.reader(f_exc)
    for row in reader:
        if row:
            include_ids.add(row[0])
print(f'Length of exclude list: {len(include_ids)}')
```

```python
import sys
csv.field_size_limit(sys.maxsize)


source_csv = './raw_data/combined_congress1879_till_2022.csv'
output_csv = './raw_data/filtered_speeches.csv'
with open(source_csv, mode='r', encoding='utf-8') as f_src, \
        open(output_csv, mode='w', encoding='utf-8', newline='') as f_out:
    
    reader = csv.reader(f_src)
    writer = csv.writer(f_out)
    ind = 0
    for row in reader:

        if ind % 10000 == 0:
            print(f'Written {ind} rows')

        if not row:
            continue
        
        row_id = row[0]
        
    
        if row_id in include_ids:
            writer.writerow(row)
            ind += 1
```
